### Defining axes, ticks, and grids

```python
x = range(1, 10)
y =[ 1, 2, 3, 4, 0, 4, 3, 2, 1]
```

This chapter is showing how to create figures based on object oriented methods.  
The process consists of 4 steps:  
- create an empty figure
- add the axes
- generate plot(s) within the figure
- Specify plotting and layout parameters for the plot(s) 

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

ax.set_xlim([1, 9])
ax.set_ylim([0, 5])
ax.set_xticks([ 1, 2, 4, 5, 6, 8, 9, 10, 11])
ax.set_yticks([0, 1, 2 , 3 , 4 ,5])

ax.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.2, .2, 1, 1])

ax.set_xlim([1, 9])
ax.set_ylim([0, 5])
ax.grid()

ax.plot(x, y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=2)
z = range(10, 20)
axes[0,0].plot(x)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(z)
```
