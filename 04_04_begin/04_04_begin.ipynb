## Segment 4 - Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10, 5
```

### Labeling plot features
#### The functional method

```python
x = range(1, 10)
y = [1, 2, 3, 4, .5, 4, 3, 2, 1 ]

plt.bar(x,y)
plt.xlabel('ypur x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1,2,3,4,7]
veh_type = ['bycicle', 'motobike', 'car', 'can', 'stroller']

plt.pie(z, labels= veh_type)
plt.show()
```

#### The object-oriented method

```python
address = '/home/d022785/SAPDevelop/git_extern/python_data_science_essential_training_part_1/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars.mpg

fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60 , fontsize='medium')
ax.set_title(' miles per gallon of Cars in mtcars Dataset')

ax.set_xlabel('car.names')
ax.set_ylabel("miles/gallon")
                    
```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

#### The object-oriented method

```python
fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60 , fontsize='medium')
ax.set_title(' miles per gallon of Cars in mtcars Dataset')

ax.set_xlabel('car.names')
ax.set_ylabel("miles/gallon")      

ax.legend(loc='best')
```

```python
fig = plt.figure()

ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60 , fontsize='medium')
ax.set_title(' miles per gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel("miles/gallon")

ax.legend(loc='best')

ax.set_ylim([0,45])

ax.annotate('Toyota Corolla', xy=(19, 33.9), xytext=(21,35), 
            arrowprops=dict(facecolor='black', shrink=0.05))
```
